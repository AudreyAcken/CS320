```python
import pandas as pd
from sklearn import preprocessing

titanic_data = pd.read_csv('../downloads/titanic_dataset.csv', sep=',')
```

```python
titanic_data.head()
```

```python
### DROP NULL DATA
titanic_homedest_data = pd.DataFrame(titanic_data['home.dest'].dropna())
titanic_homedest_data
```

```python
### CONVERT CATEGORIAL VALUES TO NUMERIC VALUES

def convert_gender(row):
    if row == 'female':
        return 1
    elif row == 'male':
        return 0
    else:
        return None
    
def convert_embarked(row):
    if row == 'S':
        return 1
    elif row == 'C':
        return 0
    elif row == 'Q':
        return 2
    else:
        return None

converted_gender = titanic_data['sex'].apply(convert_gender)
converted_embarked = titanic_data['embarked'].apply(convert_embarked)

titanic_data['embarked'] = converted_embarked
titanic_data['sex'] = converted_gender
titanic_data
```

```python
# ONE-HOT ENCODING METHOD

enc_mjob = preprocessing.LabelEncoder()
encoded_mjob = enc_mjob.fit_transform(titanic_homedest_data['home.dest'])
encoded_mjob
```
